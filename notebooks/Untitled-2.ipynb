```python
import plotly.express as px

# Script for Viewing Trends with Interactive Visualizations
# Function to plot interactive bar chart
def plot_interactive_bar(data, x, y, title, x_label, y_label):
    fig = px.bar(data, x=x, y=y, title=title)
    fig.update_layout(xaxis_title=x_label, yaxis_title=y_label)
    return fig

# Viewing trends by hour
hourly_viewings = data['Hour'].value_counts().reset_index()
hourly_viewings.columns = ['Hour', 'Viewings']
fig_hourly = plot_interactive_bar(hourly_viewings, 'Hour', 'Viewings', 
                                  'Viewing Frequency by Hour', 'Hour of the Day', 'Number of Viewings')

# Viewing trends by day
daily_viewings = data['Day'].value_counts().reset_index()
daily_viewings.columns = ['Day', 'Viewings']
fig_daily = plot_interactive_bar(daily_viewings, 'Day', 'Viewings', 
                                 'Viewing Frequency by Day', 'Day of the Month', 'Number of Viewings')

# Viewing trends by month
monthly_viewings = data['Month'].value_counts().reset_index()
monthly_viewings.columns = ['Month', 'Viewings']
fig_monthly = plot_interactive_bar(monthly_viewings, 'Month', 'Viewings', 
                                   'Viewing Frequency by Month', 'Month', 'Number of Viewings')

# Viewing trends by year
yearly_viewings = data['Year'].value_counts().reset_index()
yearly_viewings.columns = ['Year', 'Viewings']
fig_yearly = plot_interactive_bar(yearly_viewings, 'Year', 'Viewings', 
                                  'Viewing Frequency by Year', 'Year', 'Number of Viewings')

# Viewing trends by profile
profile_viewings = data['Profile Name'].value_counts().reset_index()
profile_viewings.columns = ['Profile Name', 'Viewings']
fig_profile = plot_interactive_bar(profile_viewings, 'Profile Name', 'Viewings', 
                                    'Viewing Frequency by Profile', 'Profile', 'Number of Viewings')

# Displaying one of the plots as a sample (e.g., hourly viewings)
fig_hourly.show()

```
