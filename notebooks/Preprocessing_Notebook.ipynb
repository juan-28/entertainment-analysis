```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("/Users/pranavsukumaran/Desktop/netflix/data/ViewingActivity.csv")

selected_profiles = ["Pranav", "Home", "Priya"]

# Filter the dataframe 
filtered_data = df[df['Profile Name'].isin(selected_profiles)]

```

```python
filtered_data
```

```python
# Convert 'Duration' from the format "HH:MM:SS" to timedelta
filtered_data.loc[:,'Duration'] = pd.to_timedelta(filtered_data['Duration'])

# Filter out all rows where duration is less than 5 minutes
# Make a copy of the DataFrame to ensure it's a standalone DataFrame
data_filtered = filtered_data.loc[filtered_data['Duration'] >= pd.Timedelta(minutes=5)].copy()

# Preprocess the titles and classify as TV Show or Movie
def preprocess_and_classify_title(title):
    main_title = title.split(':')[0].strip()
    content_type = 'TV Show' if 'Season' in title or 'Episode' in title or ':' in title else 'Movie'
    return main_title, content_type

# Apply the preprocessing function to the 'Title' column
titles_types = data_filtered['Title'].apply(preprocess_and_classify_title)
data_filtered.loc[:, 'Title'] = titles_types.apply(lambda x: x[0])
data_filtered.loc[:, 'Type'] = titles_types.apply(lambda x: x[1])

# Remove the unwanted columns
columns_to_remove = ['Attributes', 'Supplemental Video Type', 'Bookmark', 'Latest Bookmark']
data_filtered = data_filtered.drop(columns=columns_to_remove)

# Now your DataFrame is updated and ready for further analysis or export
data_filtered


```

```python
data_filtered['Duration'] = pd.to_timedelta(data_filtered['Duration'])
# Apply the function to the 'Duration' column
data_filtered['Duration'] = data_filtered['Duration'].dt.total_seconds() / 3600

data_filtered
```

```python

# Convert 'Start Time' to datetime
data_filtered['Start Time'] = pd.to_datetime(data_filtered['Start Time'])

# Now extract the components
data_filtered['Hour'] = data_filtered['Start Time'].dt.hour
data_filtered['Day'] = data_filtered['Start Time'].dt.day
data_filtered['Month'] = data_filtered['Start Time'].dt.month
data_filtered['Year'] = data_filtered['Start Time'].dt.year
```

```python
unique_devices = data_filtered['Device Type'].unique()

unique_devices
```

```python
def categorize_device(device_name):
    device_name = device_name.lower()  # Convert to lowercase for case-insensitive matching
    if 'iphone' in device_name or 'ipad' in device_name:
        return 'Apple Mobile Devices'
    elif 'chrome' in device_name or 'edge' in device_name or 'safari' in device_name:
        return 'Web Browsers'
    elif 'android tv' in device_name or 'smart tv' in device_name or 'tv' in device_name:
        return 'Smart TVs'
    elif 'streaming stick' in device_name or 'chromecast' in device_name or 'firetv' in device_name or 'apple tv' in device_name or 'roku' in device_name:
        return 'Streaming Devices'
    elif 'ps4' in device_name or 'ps3' in device_name:
        return 'Game Consoles'
    elif 'set top box' in device_name:
        return 'Set Top Boxes'
    else:
        return 'Other'
```

```python
# Assuming 'data_filtered' is your DataFrame and 'Device Type' is the column name
data_filtered['Device Category'] = data_filtered['Device Type'].apply(categorize_device)

# Now you can check the first few rows of your DataFrame to see the new 'Device Category' column
data_filtered
```
